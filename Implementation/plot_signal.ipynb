```python
import os
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import pywt
from scipy.ndimage import gaussian_filter

# ----------------------------
# Paths & channels
# ----------------------------
save_dir = r"E:\Collaboration Work\With Faisal\saudi conference\diagrams\wca"
os.makedirs(save_dir, exist_ok=True)

paths = {
    "Normal": r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Normal(3.0bar)\2020-04-28 18-35-35.mat",
    "MechSealScratch": r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Mechanical Seal Scratch(3.0bar)\2020-04-28 19-03-24.mat",
    "MechSealHole":    r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Mechanical Seal Hole(3.0bar)\2020-04-28 20-06-23.mat",
    "ImpellerCrack":   r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Impeller crack(3bar)\2020-04-28 16-28-33.mat",
}

channels = {
    "Normal": 0,             # Ch1
    "MechSealScratch": 2,    # Ch3
    "MechSealHole": 1,       # Ch2
    "ImpellerCrack": 0,      # Ch1
}

# ----------------------------
# Utility to load signals
# ----------------------------
def load_mat_signal(file_path, ch_idx):
    md = sio.loadmat(file_path)
    sig = md["signal"]
    fs = float(np.squeeze(md["fs"]))
    x = np.squeeze(sig[ch_idx])
    t = np.arange(x.size) / fs
    return x, fs, t

# ----------------------------
# Wavelet coherence using PyWavelets
# ----------------------------
def wavelet_coherence(x, y, fs, freqs, wavelet='cmor1.5-1.0'):
    """
    Compute wavelet coherence between x and y using PyWavelets CWT.
    Returns coherence [n_freqs, n_time].
    """
    n = min(len(x), len(y))
    x, y = x[:n], y[:n]

    # Convert frequencies to scales: scale = fs / freq
    scales = fs / freqs

    Wx, _ = pywt.cwt(x, scales, wavelet, sampling_period=1/fs)
    Wy, _ = pywt.cwt(y, scales, wavelet, sampling_period=1/fs)

    Wxy = Wx * np.conjugate(Wy)
    Sxx = np.abs(Wx) ** 2
    Syy = np.abs(Wy) ** 2

    # Smooth (Gaussian filter along both freq & time)
    S_Wxy_real = gaussian_filter(np.real(Wxy), sigma=(1, 8))
    S_Wxy_imag = gaussian_filter(np.imag(Wxy), sigma=(1, 8))
    S_Wxy = S_Wxy_real + 1j * S_Wxy_imag
    S_Sxx = gaussian_filter(Sxx, sigma=(1, 8))
    S_Syy = gaussian_filter(Syy, sigma=(1, 8))

    coh = np.abs(S_Wxy) ** 2 / (S_Sxx * S_Syy + 1e-12)
    coh = np.clip(coh, 0, 1)
    return coh

def plot_and_save_wca(coh, t, freqs, title, out_path):
    plt.figure(figsize=(12, 5))
    plt.imshow(coh[::-1, :],
               extent=[t[0], t[-1], freqs[0], freqs[-1]],
               aspect='auto', origin='lower',
               vmin=0, vmax=1, cmap='jet')
    cbar = plt.colorbar()
    #cbar.set_label("Wavelet Coherence", fontweight="bold")
    #plt.xlabel("Time [s]", fontsize=13, fontweight="bold")
    #plt.ylabel("Frequency [Hz]", fontsize=13, fontweight="bold")
    #plt.title(title, fontsize=15, fontweight="bold")
    plt.tight_layout()
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()

# ----------------------------
# Build comparisons
# ----------------------------
pairs = [
    ("Normal", "MechSealScratch"),
    ("Normal", "MechSealHole"),
    ("Normal", "ImpellerCrack"),
]

# Frequency grid
def make_freq_grid(fs, fmin=10, fmax=None, n=64):
    if fmax is None:
        fmax = fs/4
    return np.logspace(np.log10(fmin), np.log10(fmax), n)

sig_a, fs_a, t_a = load_mat_signal(paths["Normal"], channels["Normal"])

for left, right in pairs:
    sig_b, fs_b, t_b = load_mat_signal(paths[right], channels[right])

    n = min(len(sig_a), len(sig_b))
    x = sig_a[:n]
    y = sig_b[:n]
    t = t_a[:n]

    freqs = make_freq_grid(fs_a, fmin=10, fmax=fs_a/4, n=72)

    coh = wavelet_coherence(x, y, fs_a, freqs)

    out_path = os.path.join(save_dir, f"WCA_{left}_vs_{right}.png")
    title = f"WCA: {left} (Ch{channels[left]+1}) vs {right} (Ch{channels[right]+1})"
    plot_and_save_wca(coh, t, freqs, title, out_path)
    print("Saved:", out_path)

print("All WCA figures saved.")
```

```python
import os
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import pywt
from scipy.ndimage import gaussian_filter

# ----------------------------
# Output folder
# ----------------------------
save_dir = r"E:\Collaboration Work\With Faisal\saudi conference\diagrams\wca"
os.makedirs(save_dir, exist_ok=True)

# ----------------------------
# Input files
# ----------------------------
paths = {
    "Normal":            r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Normal(3.0bar)\2020-04-28 18-35-35.mat",
    "MechSealHole":      r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Mechanical Seal Hole(3.0bar)\2020-04-28 20-06-23.mat",
    "MechSealScratch":   r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Mechanical Seal Scratch(3.0bar)\2020-04-28 19-03-24.mat",
    "ImpellerCrack":     r"E:\CP Dataset\[20200428] PUMP DATA\Vibration\Impeller crack(3bar)\2020-04-28 16-28-33.mat",
}

# ----------------------------
# Channel pair for intra-file WCA (0-based)
# ----------------------------
channel_pair = (0, 1)   # Ch1 vs Ch2

# ----------------------------
# Helpers
# ----------------------------
def load_mat_channels(file_path, ch_a, ch_b):
    md = sio.loadmat(file_path)
    fs = float(np.squeeze(md["fs"]))
    sig = md["signal"]  # shape: (channels, samples)
    xa = np.squeeze(sig[ch_a])
    xb = np.squeeze(sig[ch_b])
    n = min(len(xa), len(xb))
    xa, xb = xa[:n], xb[:n]
    t = np.arange(n) / fs
    return xa, xb, fs, t

def make_freq_grid(fs, fmin=10, fmax=None, n=72):
    if fmax is None:
        fmax = fs / 4.0
    return np.logspace(np.log10(fmin), np.log10(fmax), n)

def wavelet_coherence(x, y, fs, freqs, wavelet='cmor1.5-1.0'):
    # For analytic complex Morlet: scale ~ fs / freq
    scales = fs / freqs
    Wx, _ = pywt.cwt(x, scales, wavelet, sampling_period=1/fs)
    Wy, _ = pywt.cwt(y, scales, wavelet, sampling_period=1/fs)

    Wxy = Wx * np.conjugate(Wy)
    Sxx = np.abs(Wx) ** 2
    Syy = np.abs(Wy) ** 2

    # Smooth along frequency (rows) and time (cols)
    S_Wxy_real = gaussian_filter(np.real(Wxy), sigma=(1, 8))
    S_Wxy_imag = gaussian_filter(np.imag(Wxy), sigma=(1, 8))
    S_Wxy = S_Wxy_real + 1j * S_Wxy_imag
    S_Sxx = gaussian_filter(Sxx, sigma=(1, 8))
    S_Syy = gaussian_filter(Syy, sigma=(1, 8))

    coh = np.abs(S_Wxy) ** 2 / (S_Sxx * S_Syy + 1e-12)
    return np.clip(coh, 0, 1)

def plot_wca_and_save(coh, t, freqs, out_path, title=None):
    plt.figure(figsize=(12, 5))
    im = plt.imshow(
        coh,                                     # lowhigh freq already bottomtop with origin='lower'
        extent=[t[0], t[-1], freqs.min(), freqs.max()],
        origin='lower',
        aspect='auto',
        vmin=0, vmax=1, cmap='jet'
    )
    plt.colorbar(im)  # <-- add colorbar AFTER imshow
    # Minimal look (uncomment to add labels/title)
    # plt.title(title if title else "", fontsize=15, fontweight='bold')
    # plt.xlabel("Time [s]", fontsize=12, fontweight='bold')
    # plt.ylabel("Frequency [Hz]", fontsize=12, fontweight='bold')
    plt.margins(x=0)
    plt.tight_layout()
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()

# ----------------------------
# Generate 4 images (one per condition)
# ----------------------------
for name, fpath in paths.items():
    x, y, fs, t = load_mat_channels(fpath, channel_pair[0], channel_pair[1])
    freqs = make_freq_grid(fs, fmin=10, fmax=fs/4, n=72)
    coh = wavelet_coherence(x, y, fs, freqs)

    out_file = os.path.join(save_dir, f"WCA_{name}.png")
    plot_wca_and_save(coh, t, freqs, out_file, title=f"{name} WCA Ch{channel_pair[0]+1} vs Ch{channel_pair[1]+1}")
    print("Saved:", out_file)

print("Done: 4 WCA images saved.")
```
