```python
from scipy.io import loadmat

# Load the .mat file
data = loadmat(r'E:\1 Paper Work\Cutting Tool Paper\New Dataset\600\BF_600\AE\20240814_163259_AE.mat')
```

```python
data
```

```python
# Explore the keys in the loaded data
print(data.keys())
```

```python
signals = data['signals']
fs = data['fs'][0][0]  # Extract the sampling frequency as a scalar

# Inspect the shape of the signals array
print(f"Shape of signals array: {signals.shape}")
print(f"Sampling frequency: {fs} Hz")
```

```python
import numpy as np

# Example: Normalize each signal (subtract mean, divide by standard deviation)
normalized_signals = (signals - np.mean(signals, axis=1, keepdims=True)) / np.std(signals, axis=1, keepdims=True)

# Example: Simple low-pass filter (remove high-frequency noise)
from scipy.signal import butter, filtfilt

def lowpass_filter(data, cutoff, fs, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data, axis=1)

# Apply low-pass filter with a cutoff frequency of 50 kHz
filtered_signals = lowpass_filter(normalized_signals, cutoff=50000, fs=fs)
```

```python
import matplotlib.pyplot as plt

# Plot the first signal (before and after filtering)
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(signals[0, :], label='Original Signal')
plt.title('Original Signal')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(filtered_signals[0, :], label='Filtered Signal', color='orange')
plt.title('Filtered Signal')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.fft import fft, fftfreq

# Compute the FFT of the first signal
signal_fft = fft(filtered_signals[0, :])
frequencies = fftfreq(len(signal_fft), 1/fs)

# Plot the magnitude spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(signal_fft)[:len(frequencies)//2])
plt.title('Frequency Domain Analysis')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
